```python
%matplotlib inline
```

```python
import numpy as np
import scipy.special as scsp
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,0.01)
```

```python
J0=scsp.jv(0,x)
```

```python
plt.rcParams['font.size'] = 12
plt.rcParams['font.family'] = "Serif"
```

```python
fig=plt.figure()
fsize=14
ax1=fig.add_subplot(111)

ax1.plot(x,J0,label="$J_{0}(x)$")

ax1.set_title("Bessel function $J_{0}(x)$")
ax1.set_xlabel("x",fontsize=fsize)
ax1.set_ylabel("$J_{0}(x)$",fontsize=fsize)
ax1.grid()
ax1.legend(loc=0).get_frame().set_facecolor('#FFFFFF')
```

```python
I0=scsp.iv(0,x)
K0=scsp.kv(0,x)
```

```python
fig=plt.figure()
fsize=14
ax1=fig.add_subplot(111)

ax1.plot(x,I0,label="$I_{0}(x)$")
ax1.plot(x,K0,label="$K_{0}(x)$")

ax1.set_xlim(0,1)
ax1.set_ylim(0,5)

ax1.set_title("Modified Bessel Function of order 0")
ax1.set_xlabel("x",fontsize=fsize)
ax1.set_ylabel("$I_{0}(x), K_{0}(x)$",fontsize=fsize)
ax1.grid()
ax1.legend(loc=0).get_frame().set_facecolor('#FFFFFF')
```

```python
fig=plt.figure()
fsize=14
ax1=fig.add_axes((0,0,1,0.5))
ax2=fig.add_axes((0,0.5,1,0.5),sharex=ax1)

ax1.plot(x,I0,label="$I_{0}(x)$")
ax2.plot(x,K0,label="$K_{0}(x)$")

ax1.set_xlim(0,1)
ax1.set_ylim(1,1.4)

ax2.tick_params(labelbottom="off")
```
